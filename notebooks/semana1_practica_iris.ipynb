```python
# Imports básicos
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Configuración
%matplotlib inline
sns.set_style("whitegrid")
```

## 1. Cargar Dataset

```python
# Cargar Iris
iris = load_iris()

# Convertir a DataFrame
df = pd.DataFrame(
    data=iris.data, 
    columns=iris.feature_names
)
df['species'] = iris.target
df['species_name'] = df['species'].map({
    0: 'setosa', 
    1: 'versicolor', 
    2: 'virginica'
})

# Ver primeras filas
df.head()
```

## 2. Información Básica

```python
print(f"Número de filas: {len(df)}")
print(f"Número de columnas: {df.shape[1]}")
print(f"\nColumnas: {list(df.columns)}")
```

```python
# Información del DataFrame
df.info()
```

```python
# Estadísticas descriptivas
df.describe()
```

## 3. Visualizaciones Simples

### Histograma

```python
plt.figure(figsize=(10, 5))
df['sepal length (cm)'].hist(bins=20, edgecolor='black')
plt.title('Distribución de Largo de Sépalo')
plt.xlabel('Largo (cm)')
plt.ylabel('Frecuencia')
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(8, 6))
for species in df['species_name'].unique():
    subset = df[df['species_name'] == species]
    plt.scatter(
        subset['sepal length (cm)'], 
        subset['sepal width (cm)'],
        label=species,
        alpha=0.7
    )

plt.xlabel('Largo de Sépalo (cm)')
plt.ylabel('Ancho de Sépalo (cm)')
plt.title('Sépalos por Especie')
plt.legend()
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='petal length (cm)', by='species_name')
plt.suptitle('')  # Quitar título automático
plt.title('Largo de Pétalo por Especie')
plt.ylabel('Largo (cm)')
plt.show()
```

## 4. Observaciones

Tres cosas que noté en los datos:

1. Gracias al scatterplot de los datos he podido notar que la especie de setosa tiene unas dimensiones más distintivas que las otras 2.
2. En el boxplot podemos notarlo con más obviedad, vemos que hay bastante superposición entre las especies de versiclor y virginica, lo que hace dificil distinguir entre estas 2 sabiendo unicamente las dimensiones de los pétalos.
3. Vemos que el dataset al ser un dataset ilustrativo se encuentra con datos completos, habiendo 150 registros sin datos faltantes.
